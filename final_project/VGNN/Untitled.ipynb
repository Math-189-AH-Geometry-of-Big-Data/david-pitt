```python
import pandas as pd

data_path = './data/mimic/files/'
```

```python
admissions = pd.read_csv(data_path + 'ADMISSIONS.csv')

transfers = pd.read_csv(data_path + 'TRANSFERS.csv')
```

```python
admissions
```

```python
transfers = transfers.drop(transfers[transfers['EVENTTYPE'] != 'transfer'].index)
transfers = transfers.drop(transfers[transfers['ICUSTAY_ID'].isna()].index)

transfers = transfers.drop(transfers[transfers['CURR_CAREUNIT'] == 'CSRU'].index)
transfers = transfers.drop(transfers[transfers['CURR_CAREUNIT'] == 'NICU'].index)
```

```python
transfers
```

```python
icu_admissions = transfers['HADM_ID'].tolist()

def icu_admit(hadm_id):
    return (hadm_id in icu_admissions)

admissions['ICU_ADM'] = admissions.apply(lambda row: icu_admit(row['HADM_ID']),axis=1)
```
